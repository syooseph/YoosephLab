```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm_notebook
plt.style.use('ggplot')
```

# Set up data gazing functions

## SortData()

```python
def SortData(df, print_out=False):
    """Function to sort data by magnitude and create 95% CI"""
    
    #Create a temporary column for magnitude and sort on it
    df['abs'] = abs(df.Mean)
    df.sort_values(by="abs", ascending=False, inplace=True)
    df.drop("abs", axis=1, inplace=True)
    
    #Create the 95% CI using the std and the empirical rule
    df["95_CI"] = 1.96 * df.Std
```

## CleanEdges()

```python
def CleanEdges(df, print_out=False):
    """Remove edges who have CI's spanning 0
    print_out = Boolean to print out edge name, mean, and CI for dropped edges (troubleshooting)
    creates a global variable cdf to be stored (store as copy)"""
        
    #Remove edges who have values spanning 0
    print("\nRemoving low-confidence edges")
    print(f"Starting edges: {df.shape[0]}")
    
    global cdf
    cdf = df.copy()
    
    for i in tqdm_notebook(cdf.index):
        if abs(cdf.loc[i]["Mean"]) <= abs(cdf.loc[i]["95_CI"]):
            
            #For troubleshooting purposes
            if print_out:
                print(i)
                print(f"Mean: {cdf.loc[i]['Mean']} -- CI: {cdf.loc[i]['95_CI']}")
                      
            cdf.drop(i, inplace=True)
            
    print(f"Remaining edges: {cdf.shape[0]}\n\n")
```

## ErrGraph()

```python
def ErrGraph(df0, title, n=20, save=False, out="ErrGraph", dpi=200):
    """Create bar graph using error bars of the top-n (default:20) magnitude edges using bootstrapped data
    n = amount of edges to graph
    title = graph title
    save = True to save image
    out = image name
    dpi = dpi of image
    """
    
    df = df0.head(20).copy()
    
    df.sort_values(by="Mean", inplace=True)
    
    df['labels'] = range(len(df))
    
    plt.figure(figsize=(18,18))
    sns.barplot(data=df, x=df.index, y="Mean", color='skyblue',\
                linewidth=1, edgecolor='black')

    plt.errorbar(df.index, df.Mean, yerr=df["95_CI"],\
                 fmt='none', linewidth=2, capsize=8, color='black')

    plt.axhline(0.1, color='red', linestyle="--")

    x = 0 
    for i in df.index:
        if df.loc[i].Mean > 0:
            s = int(df.loc[i]['labels'])
            
            plt.annotate(s, xy=(x, 0.012), size=14,\
                         bbox=dict(facecolor='white', zorder=2, alpha=0.65, linewidth=0.6, edgecolor='black')\
                         ,ha='center')
            x += 1

        else:
            s = int(df.loc[i]['labels'])
            plt.annotate(s, xy=(x, -0.02), size=14,\
                         bbox=dict(facecolor='white', zorder=2, alpha=0.65, linewidth=0.6, edgecolor='black'),\
                         ha='center')
            x += 1

    #Remove the x-ticks for neater image
    plt.tick_params(axis='x', bottom=False, labelbottom=False)
    
    plt.ylabel("Mean Edge-weight\n", size=15)

    #Set the numeric legend
    leg = {}
    for i in df.index:
        label = int(df.loc[i]["labels"])
        leg[label] = i

    s = """"""
    cnt = 0
    for i in leg:
        if i == 0:
            s += str(i) + ". " + leg[i].replace(" <--> ", "<>")
        else:   
            s += "\n" + str(i) + ". " + leg[i].replace(" <--> ", "<>")

    plt.text(x=-1, y=0.37, s=s, size=14,\
             bbox=dict(facecolor="white", linewidth=1, edgecolor='black'))
    
    
   #Quick check for neatness; if no negative values, no need for axhline at -0.1 and note must fit @ bottom
    if df.Mean.min() > 0:
        y2 = -0.1
        plt.axhline(-0.1, color='none', linestyle="none")
    else:
        y2 = df.Mean.min() - 0.075
        plt.axhline(-0.1, color='red', linestyle="--")
        
    plt.text(x=15, y=y2, s="""Iterations: 1000
Error bars: 95% CI """, size=20,\
            bbox=dict(facecolor="wheat", linewidth=1, edgecolor='k'))
    
    plt.title("Top {} mean edge-weights by magnitude\n {}".format(n, title))

    plt.tight_layout()
    
    if save:
        plt.savefig(out, dpi=dpi)

    plt.show()
```

## EdgeHist()

```python
def EdgeHist(df, title, save=False, out="EdgeHist.png", dpi=200):
    """Create edge weight histogram
    title = title for histogram
    save = True to save image
    out = image name for saving
    dpi = dpi for image"""
    
    plt.figure(figsize=(10,8))

    sns.distplot(df.Mean, kde=False, color='gold', hist_kws=dict(edgecolor="k", linewidth=1, alpha=1))
    
    
    #Useful for fitting anootation to the graph neatly
    xmin, xmax, ymin, ymax  = plt.axis()


    plt.text(df.Mean.max()-0.01, ymax*0.1, "Max\n{}"\
             .format(round(df.Mean.max(), 3)),\
             horizontalalignment= "center", bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)

    plt.text(df.Mean.min()+0.02, ymax*0.1, "Min\n{}".format(round(df.Mean.min(), 3)),\
             horizontalalignment= "center", bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)



    #Plot out the amount of samples in various bins
    plt.axvline(x=0.1, color='red', linestyle="--")
    txt = len(df[(df.Mean < 0.1) & (df.Mean > -0.1)])
    plt.text(-0.01, ymax*0.8, "n={}".format(txt), horizontalalignment= "center",\
             bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)
    
    plt.axvline(x=0.2, color='red', linestyle="--")
    txt = len(df[(df.Mean > 0.1) & (df.Mean <0.2)])
    plt.text(0.15, ymax*0.6, "n={}".format(txt), horizontalalignment= "center",\
             bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)


    plt.axvline(x=0.3, color='red', linestyle="--")
    txt = len(df[(df.Mean > 0.2) & (df.Mean < 0.3)])
    plt.text(0.25, ymax*0.4, "n={}".format(txt), horizontalalignment= "center",\
             bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)

    txt = len(df[(df.Mean > 0.3)])
    plt.text(0.35, ymax*0.2, "n={}".format(txt), horizontalalignment= "center",\
             bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)


    #Now plot out the negatives
    plt.axvline(x=-0.1, color='red', linestyle="--")
    txt = len(df[(df.Mean < -0.1) & (df.Mean > -0.2)])
    plt.text(-0.15, ymax*0.6, "n={}".format(txt), horizontalalignment= "center",\
             bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)

    plt.axvline(x=-0.2, color='red', linestyle="--")
    txt = len(df[(df.Mean < -0.2) & (df.Mean > -0.3)])
    plt.text(-0.25, ymax*0.4, "n={}".format(txt), horizontalalignment= "center",\
             bbox={"facecolor":"beige", "edgecolor":"black"}, size=8)


    plt.title("""Edge-weight Distribution
    {}
    Edges: {}    Mean: {}    Median: {}"""\
              .format(title, len(df.index), round(df.Mean.mean(), 4), round(df.Mean.median(),4)))

    plt.xlabel("\nWeights")
    plt.ylabel("Edges\n")

    plt.xticks(np.arange(-0.6, 0.65, step=0.1))

    plt.tight_layout()
    
    if save:
        plt.savefig(out, dpi=dpi)
    
    plt.show()
```

## ViewSparse()

```python
def ViewSparse(df):
    df["abs"] = abs(df.Mean)
    df.sort_values(by="abs", ascending=False, inplace=True)

    sp70 = int(len(df) * 0.3)
    sp80 = int(len(df) * 0.2)
    sp90 = int(len(df) * 0.1)
    sp95 = int(len(df) * 0.05)

    print("Mean Edge-weight")

    print("\nEdges retained at 70%: {}".format(sp70))
    print(">>>>>>> Lowest absolute association at 70% sparsity <<<<<<<")
    print(df.Mean.head(sp70).tail(1))

    print("\n\nEdges retained at 80%: {}".format(sp80))
    print(">>>>>>> Lowest absolute association at 80% sparsity <<<<<<<")
    print(df.Mean.head(sp80).tail(1))

    print("\n\nEdges retained at 90%: {}".format(sp90))
    print(">>>>>>> Lowest absolute association at 90% sparsity <<<<<<<")
    print(df.Mean.head(sp90).tail(1))
    
    print("\n\nEdges retained at 95%: {}".format(sp95))
    print(">>>>>>> Lowest absolute association at 95% sparsity <<<<<<<")
    print(df.Mean.head(sp95).tail(1))
    
    df.drop('abs', inplace=True, axis=1)
```

# Set up functions for overlap analysis

## FindOvers()

```python
def FindOvers(df1, df2, thresh, df1_name, df2_name, print_out=False):
    """Find overlapping edges using 95_CI's, if the edges do not overlap, it will be given a 0 in one group
    outputs tdf: df containing the total edges (overlapping/non-overlapping)
    thresh = edge_weight threshold
    df1_name = name to display for df1
    df2_name = name to display for df2
    print_out = boolean to print the top 20 overlaps
    """
    ht = df1[abs(df1.Mean) >= thresh].rename(columns={"Mean": df1_name + "_Mean","Std": df1_name + "_Std",\
                                                    "95_CI":df1_name + "_95_CI"})
    kt = df2[abs(df2.Mean) >= thresh].rename(columns={"Mean": df2_name + "_Mean", "Std": df2_name + "_Std",\
                                                    "95_CI":df2_name + "_95_CI"})
    
    
    global tdf
    tdf = pd.concat([kt, ht], axis=1, sort=True).fillna(0)
    tdf["AbsDiff"] = tdf.apply(lambda x:abs(x[df1_name + "_Mean"] - x[df2_name + "_Mean"]), axis=1)
    tdf.sort_values(by="AbsDiff", inplace=True, ascending=True)
    
    #Create dictionaries to store the values of the range for df1 and for df2
    global d1
    d1 = {}
    global d2
    d2 = {}
    
    for i in tqdm_notebook(tdf.index, desc="Calculating upper/lower bounds"):
        d1[i] = [tdf.loc[i][df1_name + "_Mean"] - tdf.loc[i][df1_name+"_95_CI"],\
                  tdf.loc[i][df1_name + "_Mean"] + tdf.loc[i][df1_name+"_95_CI"]]
        
        d2[i] = [tdf.loc[i][df2_name + "_Mean"] - tdf.loc[i][df2_name+"_95_CI"],\
                  tdf.loc[i][df2_name + "_Mean"] + tdf.loc[i][df2_name+"_95_CI"]]
        
        
    #If the df2_lower_bound or the df2_upper_bound is between
        #the df1_lower_bound and the df1_upper bound, then the ranges overlap
        
    global sigs
    sigs = {} #Dict to store whether a value is significantly different using the 95% CI's
    
    for i in tqdm_notebook(tdf.index, desc="Calculating significance"):
        if d1[i][0] <= d2[i][0] <= d1[i][1]:
            sigs[i] = False
        elif d1[i][0] <= d2[i][1] <= d1[i][1]:
            sigs[i] = False
        elif d2[i][0] <= d1[i][0] <= d2[i][1]:
            sigs[i] = False
        elif d2[i][0] <= d1[i][1] <= d2[i][1]:
            sigs[i] = False
        
        else:
            sigs[i] = True
            
    
    tdf["Sig"] = list(sigs.values())
    
    
    overlaps = tdf[tdf.Sig==False].shape[0]
    print(f"There are {overlaps} ovrelapping edges @ {thresh} edge-weight threshold")
    print(f"There are {tdf[tdf.Sig].shape[0]} significantly different edges\n")
    
    print(f"The average absolute difference between {df1_name} and \
{df2_name} = {round(tdf[tdf.Sig==False].AbsDiff.mean(), 5)}")
    
    if print_out:
        print("\nTop 20 overlapping edges")
        print(tdf[tdf.Sig==False][[df1_name + "_Mean", df2_name + "_Mean"]].head(20))
```

# Examine overlaps in the context of replacement

## Create function to find overlaps with replacement

### FindOverReps()

```python
def FindOverReps(df1, df2, df1_name, df2_name, thresh):
    """Function to find overlaps with replacement
    thresh = edge-weight threshold"""
    
    pcnt=0
    
    for e1 in df1[df1.Mean > thresh].index:
        b11,b12 = e1.split(" <--> ")
        
        for e2 in df2[df2.Mean > thresh].index:
            b21,b22 = e2.split(" <--> ")
            
            if b11 == b21 and b12 != b22:
                g12 = b12.split("_")[0]
                
                if g12 in b22:
                    print(f"{df1_name}: {b11} :-: {b12} -- {round(df1.loc[e1].Mean, 3)}")
                    print(f"{df2_name}: {b21} :-: {b22} -- {round(df2.loc[e2].Mean, 3)}\n")
                    pcnt+=1
                    
            elif b11 != b21 and b12 == b22:
                g11 = b11.split("_")[0]
                
                if g11 in b21:
                    print(f"{df1_name}: {b11} :-: {b12} -- {round(df1.loc[e1].Mean, 3)}")
                    print(f"{df2_name}: {b21} :-: {b22} -- {round(df2.loc[e2].Mean, 3)}\n")  
                    pcnt+=1
                    
            elif b21 == b11 and b22 != b12:
                g22 = b22.split("_")[0]
                
                if g22 in b12:
                    print(f"{df1_name}: {b11} :-: {b12} -- {round(df1.loc[e1].Mean, 3)}")
                    print(f"{df2_name}: {b21} :-: {b22} -- {round(df2.loc[e2].Mean, 3)}\n")
                    pcnt+=1
                    
            elif b21 != b11 and b22 == b12:
                g21 = b21.split("_")[0]
                
                if g21 in b11:
                    print(f"{df1_name}: {b11} :-: {b12} -- {round(df1.loc[e1].Mean, 3)}")
                    print(f"{df2_name}: {b21} :-: {b22} -- {round(df2.loc[e2].Mean, 3)}\n")  
                    pcnt+=1
            
    print(f"There are {pcnt} partial overlaps")
```
